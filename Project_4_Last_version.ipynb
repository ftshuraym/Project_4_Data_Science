# System recommender function 

```python
def build_chart(data, percentile=.5):
    
#Ask for preferred genres
    print("Input preferred generes") # done 
    generes = input()
    
    #Ask for lower limit of price
    print("Input  lower price")
    lower_price = float(input())
    
    #Ask for upper limit of price
    print("Input higher price")
    high_price = int(input())
    
    
    #Define a new books variable to store the preferred books. Copy the contents of gen_df to books
    books = data.copy()

    #Filter based on the condition
    books = data[(data['generes'] == generes) & 
                    (data['price'] >= lower_price) & 
                    (data['price'] <= high_price)]
    
    #Compute the values of C and m for the filtered books
    C = data['rating'].mean()
    m = data['voters'].quantile(percentile)
    
    #Only consider books that have higher than m votes. Save this in a new dataframe q_books
    q_books = books.copy().loc[data['voters'] >= m]
    
    #Calculate score using the IMDB formula
    data['score'] = data.apply(lambda x: (x['voters']/(x['voters']+m) * x['rating'])+ (m/(m+x['voters']) * C)
                                      ,axis=1)

    #Sort books in descending order of their scores
    q_books = q_books.sort_values('score', ascending=False)
    
    return q_books
```

```python
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import TfidfVectorizer 
```

```python
# copying the data 
df= data.copy()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# using TfidfVectorizer
books_tfidf = TfidfVectorizer(stop_words='english')
```

```python
# fit and transform 'title'
book_description_matrix = books_tfidf.fit_transform(df['title'])
```

```python
book_description_matrix.shape
```

```python
# checking type of matrix 
book_description_matrix
```

```python
# Calculating similarity   
cosine_similarity = linear_kernel(book_description_matrix, book_description_matrix)
```

```python
# taking 'Title' indexes.
indices = pd.Series(df['title'].index)
```

```python
def recommend(index, cosine_sim=cosine_similarity):
    id = indices[index]
    
    # Get the pairwsie similarity scores of all books compared to that book, 
    # sorting them and getting top 5
    
    similarity_scores = list(enumerate(cosine_sim[id]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores = similarity_scores[1:6]

    # Get the books index
    books_index = [i[0] for i in similarity_scores]

    # Return the top 5 most similar books using integer-location based indexing (iloc)
    return df['title'].iloc[books_index]
```
